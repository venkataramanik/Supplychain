```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Generate 20 weeks of demand data
weeks = pd.date_range(start='2024-01-01', periods=20, freq='W-MON')
demand = np.random.randint(100, 300, size=20)

# Create dataframe
df = pd.DataFrame({'Week': weeks, 'Demand': demand})
df.head()
# Rolling average (window = 4 weeks)
df['Rolling_Avg'] = df['Demand'].rolling(window=4).mean()

# Projected growth (10%)
df['Projected'] = df['Demand'] * 1.10

df.tail()

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(df['Week'], df['Demand'], label='Actual Demand', marker='o')
plt.plot(df['Week'], df['Rolling_Avg'], label='4-Week Rolling Avg', linestyle='--')
plt.plot(df['Week'], df['Projected'], label='Projected (+10%)', linestyle=':')

plt.title('Simple Demand Forecast')
plt.xlabel('Week')
plt.ylabel('Units')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

# 📦 Project 1 – Simple Demand Forecast

This project simulates a 20-week demand forecast using basic statistical techniques:
- **Rolling Average Forecast** (4-week window)
- **Projected Growth Forecast** (+10%)

It uses Python, Pandas, and Matplotlib to visualize actual vs forecasted demand, suitable for entry-level demand planning or e-commerce use cases.
